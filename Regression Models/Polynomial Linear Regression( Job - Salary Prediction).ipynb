```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 
```

```python
data = pd.read_csv(r"C:\Users\AKASH\Desktop\ML\Machine Learning A-Z (Codes and Datasets)\Regression\Section 6 - Polynomial Regression\Python\Position_Salaries.csv")
```

```python
data.head()
```

Plolynomial Linear Regregion is a input/output relation between the linear combinations of the powers of the single feature
y=f(x)=β0+β1x+β2x2+β3x3+…

```python
# lets first train a linear model on the single feature 
X = data["Level"]
y = data["Salary"]
X = np.array(X).reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm =  LinearRegression()
```

```python
lm.fit(X,y)
```

```python
# Now lets train it to the polynomial Linear regression model 
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2) # for now we are putting the value of (n=2) 

# now we should transform the feature x into mutiple features (x,x^2,x^3....,x^n)
X_poly = poly_reg.fit_transform(X)
```

```python
X_poly
```

```python
lm2 = LinearRegression()
```

```python
lm2.fit(X_poly,y)
```

```python
# Visualising the Linear Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, lm.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Polynomial Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, lm2.predict(X_poly), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Predicting a new salary with linear regression 
lm.predict([[6.5]])  # it is going to give a very much wierd ammount of money
```

```python
# Predicting a new salary with Polynomial regression 
lm2.predict(poly_reg.fit_transform([[6.5]]))
```
